# Baixando e carregando os dados - portal da transparência

```python
caminho_dados = '/content/drive/MyDrive/Analise de Dados - PYTHON/2023_Viagem.csv'
```

```python
import pandas as pd

pd.set_option('display.max_columns', None) #para exibir todas as colunas da tabela
pd.set_option('display.float_format', '{:.2f}'.format) #para converter as colunas de números que estavam como string

#Lendo os dados corretamente - #IMPORTANTE: os dados do portal da transparencia estão em outro padrão,
#então precisa indicar isso para o pandas, o enconding e a separação por ;
df_viagens = pd.read_csv(caminho_dados, encoding='Windows-1252', sep=';', decimal=',')


```

```python
df_viagens.head()
```

# Contando frequência - agrupamento de dados

```python
#Criando nova coluna de despesas
df_viagens['Despesas'] = df_viagens['Valor diárias'] + df_viagens['Valor passagens'] + df_viagens['Valor outros gastos']

df_viagens.head()
```

```python
#Como analisar a proporção de viagens que cada cargo público fez por ano:

df_viagens['Cargo'].value_counts() #contar quantas vezes aparece o cargo na tabela - frequência
(df_viagens['Cargo'].value_counts(normalize=True) * 100).rename('Proporção de viagens').reset_index()
#tornar os dados proporcionais (em porcetagem, a multiplicação torna a porcentagem sem o 0 antes.
#O rename altera o nome depois de 'cargo' e o reset' formata os titulos)
```

```python
#Agrupando os dados a partir da coluna Cargo

df_viagens.groupby('Cargo')['Despesas'].sum()
#groupby agrupa os dados, após isso precisar informar qual a proxima coluna a ser trabalhada, o sum() vai somar a coluna despea para o cargo
# se quiser a média de gastos por cargo ao invés da soma é só usar o .mean() no lugar de sum()
#para saber a viagem mais cara por cargo substituir por .max()

df_viagens.groupby('Cargo')['Despesas'].sum().reset_index().sort_values(by="Despesas", ascending=False)
#sort - utilizado para ordenar os valores e a sequencia que queremos
```

# Séries booleanas e filtros

```python
#Colocando as informações da tabela dentro de uma variável para acessar os dados
viagens_por_cargo = (df_viagens['Cargo'].value_counts(normalize=True) * 100).rename('Proporção de viagens').reset_index()

#Comparação em python - utilizando booleanos para filtras os dados
filtro_mais_de_1_pct = viagens_por_cargo['Proporção de viagens'] > 1
viagens_por_cargo[filtro_mais_de_1_pct]
```

Criando um novo filtro de acordo com o cargo

```python
filtro_cargo_tecnico = viagens_por_cargo['Cargo'].str.startswith("TECNICO")
viagens_por_cargo[filtro_cargo_tecnico]

#Juntando Filtros
viagens_por_cargo[filtro_cargo_tecnico & filtro_mais_de_1_pct]
```

```python
#Teste - pegar as viagens que gastaram mais de 10 milhoes

soma_despesas = df_viagens.groupby('Cargo')['Despesas'].sum().reset_index()
filtro_teste = soma_despesas['Despesas'] > 10_000_000.00

soma_despesas[filtro_teste]

#forma mais comum de fazer:
#soma_despesas[soma_despesas['Despesas'] > 10_000_000]
```

# Trabalhando com Datas

```python
df_viagens[['Período - Data de início', 'Período - Data de fim']]
```

```python
# convertendo para formato de data e especificando que está no formato brasileiro - alterar na tabela principal
df_viagens['Período - Data de início'] = pd.to_datetime(df_viagens['Período - Data de início'], format='%d/%m/%Y')
df_viagens['Período - Data de fim'] = pd.to_datetime(df_viagens['Período - Data de fim'], format='%d/%m/%Y')
```

Observações possíveis de se fazer com as datas:

```python
#Mostrar o nome dos meses e criar uma variável para guardar essa informação
df_viagens['Mês da Viagem'] = df_viagens['Período - Data de início'].dt.month_name()
```

```python
#Para saber quanto tempo durou a viagem - em dias
df_viagens['Dias de Viagem'] = (df_viagens['Período - Data de fim'] - df_viagens['Período - Data de início']).dt.days
```

# Combinando agregrações e filtros

```python
#Construindo uma tabela inteira nova com meus dados filtrados

(df_viagens
 .groupby('Cargo')
 .agg(
     despesa_media=('Despesas', 'mean'),
     duracao_media=('Dias de Viagem', 'mean'),
     despesas_totais = ('Despesas', 'sum'),
     destino_mais_frequente = ('Destinos', pd.Series.mode),
     n_viagens=('Nome', 'count')
 )
 .reset_index()
 #.sort_values(by='despesas_totais', ascending=False)
     )
```

```python
#Criando uma variável para guardar a planilha nova

df_viagens_consolidado = (
    df_viagens
   .groupby('Cargo')
   .agg(
     despesa_media=('Despesas', 'mean'),
     duracao_media=('Dias de Viagem', 'mean'),
     despesas_totais = ('Despesas', 'sum'),
     destino_mais_frequente = ('Destinos', pd.Series.mode),
     n_viagens=('Nome', 'count')
      )
   .reset_index()
 #.sort_values(by='despesas_totais', ascending=False)
     )
```

```python
#criando uma variável para filtro de proporção

df_cargos = df_viagens['Cargo'].value_counts(normalize=True).reset_index()

df_cargos.loc[df_cargos['proportion'] > 0.01, 'Cargo']

#Colocando tudo em uma nova variável
cargos_relevantes = df_cargos.loc[df_cargos['proportion'] > 0.01, 'Cargo']
```

Quais viagens consolidadas estão nos cargos relevantes?

```python
filtro = df_viagens_consolidado['Cargo'].isin(cargos_relevantes)

df_final = df_viagens_consolidado[filtro]
```

# Criando e customizando um gráfico

```python
#Exemplo de Gráficos no pandas - limitado
df_final = df_final.sort_values(by='n_viagens', ascending=False)
df_final.plot(x='Cargo', y='n_viagens', kind='bar')
```

```python
#Gráficos mais robustos utilizando outra biblioteca

import matplotlib.pyplot as plt
caminho_figura2 = '/content/drive/MyDrive/Analise de Dados - PYTHON/output/figura2.png'
#Criando uma figura com um gráfico central e demilitando o tamanho

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(df_final['Cargo'], df_final['n_viagens'], color='#127817')
ax.invert_yaxis() #invertendo a ordem das barras
ax.set_facecolor('#202423') #alterando a cor do fundo do gráfico

fig.suptitle('Viagens por cargo público (2023)')
plt.figtext(0.67, 0.89, 'Fonte: Portal da Transparência', fontsize=8)

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.yticks(fontsize=8)

plt.xlabel('Número de viagens')
plt.ylabel('Cargo')

plt.savefig(caminho_figura2, bbox_inches='tight')
plt.show()
```
